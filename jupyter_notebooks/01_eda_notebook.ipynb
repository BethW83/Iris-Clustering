Next: visualise the data
- species bar chart
- violin plots
- heatmap (seaborn)

```python
# bar plot to show the distribution of the different Iris species in the dataset
sns.countplot(x="Species", data=df)
plt.title("Distribution of Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
# violin plots to visualize the distribution of each feature by species
# iterate through each feature and create a violin plot for each one

sns.set_style("whitegrid")

for feature in df.columns[1:-1]:  # skip Id and Species
    plt.figure(figsize=(10, 6))
    sns.violinplot(x="Species", y=feature, data=df)
    plt.title(f"Distribution of {feature} by Species")
    plt.tight_layout()
    plt.show()
```

```python
# create a heatmap to visualize the correlation between the features

# remove species column as we only want to look at the correlation between the numeric features
corr = df.select_dtypes(include="number").corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="cool", linewidths=0.5)
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```
